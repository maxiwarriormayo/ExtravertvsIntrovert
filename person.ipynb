```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# joining the two datasets together
person1 = pd.read_csv('personality_datasert.csv')
person2 = pd.read_csv('personality_dataset.csv')
person = pd.concat([person1,person2], axis = 0)


#cleaning the data 
person = person.dropna()
    
# confirming whether null values are present 

person.info()
```

Insights Present

```python
#Intraversion VS Extraversion
total_time_spent_lone = person.groupby('Personality')['Time_spent_Alone'].mean().reset_index()
stage_fear = person.groupby('Personality')['Stage_fear'].value_counts().reset_index()
#on further evaluation doing analysis by just reading value couts does not provide full truths

#shows no real difference in fear between introverts and extroverts but shows no correlation between intraversion and Stage_fear


average_friend_size = person.groupby('Personality')['Friends_circle_size'].mean().reset_index()
average_friend_size
# shows that introverts and extroverts have a large discrepancy in terms of friend size. seeing as introverts usually are in groups of 3 unlike extroverts in groups of 9

print(total_time_spent_lone,stage_fear)
```

```python
#extrovert relation of going outside and post frequency

extra_gp = person[person.Personality == 'Extrovert']
intra_gp = person[person.Personality == 'Introvert']
```

```python

plt.scatter(intra_gp.Going_outside, intra_gp.Post_frequency, color = 'Blue',label = 'Introverts',alpha=0.1, s=100)

plt.scatter(extra_gp.Going_outside, extra_gp.Post_frequency, color = 'Green', label = 'Extroverts',alpha=0.1, s=100)


# Shows high correlation between going outside and how much posts and shows that introverts do more of the gping out
plt.xlabel("How long outside(in Hours)")
plt.ylabel("How many posts")

plt.title('Correlation between Going Outside and Post Frequency')
plt.grid(True)

plt.legend()
plt.show()
#people who go out longer post more frquently plus more extroverts go out longer and post more frequently in comparison to introverts
```
